```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.cluster import KMeans
import warnings
warnings.filterwarnings("ignore", category=UserWarning, module="sklearn.cluster._kmeans")
from sklearn.linear_model import LogisticRegression
from sklearn.svm import SVC
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score, mean_squared_error
from sklearn.decomposition import PCA
```

```python
data = data.drop(columns=['id'], errors='ignore')
```

```python
X = data.drop(columns=['diagnosis']) 
y = data['diagnosis']
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
```

```python
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)
```

```python
kmeans = KMeans(n_clusters=2, random_state=42)
kmeans.fit(X_train_scaled)
```

```python
X_pca = PCA(n_components=2).fit_transform(X_train_scaled)

plt.figure(figsize=(8, 6))
plt.scatter(X_pca[:, 0], X_pca[:, 1], c=kmeans.labels_, cmap='coolwarm', s=50, alpha=0.7)
plt.title('K-Means Clustering (K=2)')
plt.xlabel('PCA Component 1')
plt.ylabel('PCA Component 2')
plt.colorbar(label='Cluster')
plt.savefig("K-Means.png", format='png', dpi=300, bbox_inches='tight')
plt.show()
```

```python
### Logistic Regression 
```

```python
log_reg = LogisticRegression(max_iter=1500, C=0.5, random_state=42)
log_reg.fit(X_train_scaled, y_train)
y_pred_log_reg = log_reg.predict(X_test_scaled)
```

```python
svc = SVC(kernel='rbf', C=2, gamma='scale', random_state=42)
svc.fit(X_train_scaled, y_train)
y_pred_svc = svc.predict(X_test_scaled)
```

```python
rf = RandomForestClassifier(n_estimators=200, max_depth=10, random_state=42)
rf.fit(X_train_scaled, y_train)
y_pred_rf = rf.predict(X_test_scaled)
```

```python
log_reg_acc = accuracy_score(y_test, y_pred_log_reg)
svc_acc = accuracy_score(y_test, y_pred_svc)
rf_acc = accuracy_score(y_test, y_pred_rf)
print("Logistic Regression: ", log_reg_acc)
print("SVC: ", svc_acc)
print("Random Forest: ", rf_acc )
```

```python
accuracies = {
    'Logistic Regression': log_reg_acc,
    'SVC': svc_acc,
    'Random Forest': rf_acc
}

plt.figure(figsize=(8, 6))
sns.barplot(x=list(accuracies.keys()), y=list(accuracies.values()))
plt.title('Comparison of Model Accuracies')
plt.ylabel('Accuracy')
plt.ylim(0.8, 1.0)
plt.savefig("comaprision of Model Accuries.png", format='png', dpi=300, bbox_inches='tight')
plt.show()
```

# Part 3 

```python
from sklearn.linear_model import Lasso, Ridge
from xgboost import XGBRegressor
from sklearn.metrics import mean_squared_error
```

```python
label_column = 'area_mean'
```

```python
X = data.drop(columns=[label_column, 'diagnosis'])  
y = data[label_column]  
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
```

```python
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)
```

```python
lasso = Lasso(alpha=0.1, max_iter=1000, random_state=42)
lasso.fit(X_train_scaled, y_train)
y_pred_lasso = lasso.predict(X_test_scaled)
mse_lasso = mean_squared_error(y_test, y_pred_lasso)
```

```python
ridge = Ridge(alpha=1.0, max_iter=1000, random_state=42)
ridge.fit(X_train_scaled, y_train)
y_pred_ridge = ridge.predict(X_test_scaled)
mse_ridge = mean_squared_error(y_test, y_pred_ridge)
```

```python
xgb = XGBRegressor(n_estimators=200, learning_rate=0.05, max_depth=5, random_state=42)
xgb.fit(X_train_scaled, y_train)
y_pred_xgb = xgb.predict(X_test_scaled)
mse_xgb = mean_squared_error(y_test, y_pred_xgb)
```

```python
mse_values = {
    'Lasso Regression': mse_lasso,
    'Ridge Regression': mse_ridge,
    'XGBoost Regressor': mse_xgb
}
```

```python
plt.figure(figsize=(8, 6))
sns.barplot(x=list(mse_values.keys()), y=list(mse_values.values()), color='skyblue')
plt.title('Comparison of Mean Squared Errors')
plt.ylabel('Mean Squared Error (MSE)')
plt.show()
```

```python
print("Mean Squared Errors:")
for model_name, mse in mse_values.items():
    print(f"{model_name}: {mse:.4f}")
```
